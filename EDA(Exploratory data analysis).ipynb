**What is bivariate analysis?** Bivariate analysis is a statistical method examining how two different things are related.
The bivariate analysis aims to determine if there is a statistical link between the two variables and, if so,
how strong and in which direction that link is.

# 1. Scatterplot (Numerical - Numerical)

```python
#scatterplot is basically for numerical data if both column are numerical and we want to compare it
sns.scatterplot(x=tips['total_bill'],y=tips['tip'])
```

```python
#we will add hidden parameter called "hue" that which customer are male and female
#this is example of multivariet analysis
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'])
```

```python
#we will add hidden parameter called "style" that which customer are smoker and no-smoker
#this is also example of multivariet analysis
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'])
```

# 2. Bar plot (Numerical - Categorical)

```python
#on x-axis their is categorical column and on y-axis their is numerical column
sns.barplot(x=titanic_dataset['Pclass'],y=titanic_dataset['Age'])
```

```python
sns.barplot(x=titanic_dataset['Pclass'],y=titanic_dataset['Age'],hue=titanic_dataset['Sex'])
```

# 3. Distplot(Numerical - Categorical)

```python
#blue graph represent "death" orange graph represent "live"
ax = plt.subplots(figsize=(6, 3))
sns.distplot(titanic_dataset[titanic_dataset['Survived']==0]['Age'],hist=False)
sns.distplot(titanic_dataset[titanic_dataset['Survived']==1]['Age'],hist=False)
```

# 4. HeatMap(Categorical - Categorical)

```python
#0=death 1=lived
"""
plcass=1 80 dead 136 lived
plcass=2 97 dead 87 lived
plcass=3 372 dead 119 lived
"""
pd.crosstab(titanic_dataset['Pclass'],titanic_dataset['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic_dataset['Pclass'],titanic_dataset['Survived']))
```
